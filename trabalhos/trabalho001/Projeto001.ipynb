## Introdução

A **Agnar** está planejando otimizar suas estratégias de marketing para o próximo feriadão. O time de marketing da marca coletou dados para os seus 5 produtos mais vendidos. Os dados coletados estão em grandes silos, ou seja, não possuem conexão aparente e, em sua criação, foram mantidos somente com seus "donos". Dados em silos não podem entregam todo o seu potencial e o time de marketing não tem as capacidade técnicas necessárias para executar análise quantitativas e qualitativas desses dados. Eles convidaram um Cientista de Dados (VOCÊ) para resolver esse desafio por eles e auxiliar na concepção de estratégias para otimizar a gestão dos inventários das lojas e garantir que a disponibilidade de produtos durante a alta de vendas nos feriados.

Você terá **20 minutos!** para apresentar os insights da sua análise e propostas para o cliente. Na reunião de apresentação estarão o Líder de Vendas e a Líder da Logística. Eles irão avaliar se seus insights e soluções estão aderentes a realidade de cada time e os possíveis impactos de cada insight.

## Roteiro para o sucesso do trabalho

* Data Wrangling : transorme e limpe seus dados, crie uma ou mais bases que sejam ótimas para um trabalho de análise de dados.
* Storytelling : utilize seus dados para contar uma história ( [O que é Data Storytelling?](https://paulovasconcellos.com.br/o-que-%C3%A9-data-storytelling-ac5a924dcdaf) )
* Diversifique : colete mais dados, existem dados abertos que podem ser de extremo valor para todo problema, ele só precisa ser encontrado.
* Análise de dados : utilize ferramentas estatísticas, experiência e conhecimentos adquiridos através de pesquisa para direcionar sua análise e sempre se pergunte "Os insights entregados podem ser convertidos em ação? Qual o impacto dessa informação? Como transformar ela em insight?"
* Conhecimento de Negócio : pesquise e entenda como funciona o negócio que você está analisando (dica: nesse caso pesquisa sobre estoque mínimo, máximo e de segurança)
* Concisão e Línguagem : Lembre-se que você está apresentando para um público não técnico e que mostrar código não é a melhor forma de convencer esse público, seja conciso e direto.
* Métricas e Medidas : crie métricas de acompanhamento que sejam simples e demonstrem a magnitude do problema e também a magnitude da solução.
